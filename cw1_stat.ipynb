```python
from scipy import stats
import seaborn as sns
df = pd.read_csv('refugee_data.csv')
```

```python
df.drop(columns=['Year']).describe()
```

```python
sns.heatmap(df.drop(columns=['Year']).corr())
```

```python
skewness = stats.skew(df['Stateless persons'])
kurt = stats.kurtosis(df['Stateless persons'])
mode = stats.mode(df['Stateless persons'])

print(f"Skośność: {skewness}")
print(f"Kurtoza: {kurt}")
print(f"Dominanta: {mode.mode[0]}")
```

```python
df = pd.read_csv('brain_size.csv', delimiter=';')
df=df.drop(columns=['Unnamed: 0'])
df
```

```python
import matplotlib.pyplot as plt

df[['VIQ', 'PIQ', 'FSIQ']].hist()
plt.suptitle('Histogramy VIQ, PIQ, FSIQ')
plt.show()
```
